***Задание 1***

Напишите функцию date_range, которая возвращает список дней между датами start_date и end_date. Даты должны вводиться в формате YYYY-MM-DD.

```python
from datetime import datetime
from datetime import timedelta
```

```python
def date_range(start_date,end_date):
    cur_day = datetime.strptime(start_date,'%Y-%m-%d')
    end_date = datetime.strptime(end_date,'%Y-%m-%d')
    list_days = []
    
    while cur_day <= end_date:
        list_day = cur_day.day
        list_days.append(list_day)
        cur_day += timedelta(days = 1)
    return list_days
        
date_range('2018-09-12','2018-10-12')
```

Дополните функцию из первого задания проверкой на корректность дат. В случае неверного формата или если start_date > end_date должен возвращаться пустой список.

```python
# в моём случае он и так пустой возвращается :)
date_range('2018-11-12','2018-10-12')
```

```python
# и лучше тогда не пустой список возвращать, а сообщение
def date_range_2(start_date,end_date):
    cur_day = datetime.strptime(start_date,'%Y-%m-%d')
    end_date = datetime.strptime(end_date,'%Y-%m-%d')
    list_days = []
    if cur_day > end_date:
        return 'Не правильно введены даты, начальная больше конечной'
    else:
        while cur_day <= end_date:
            list_day = cur_day.day
            list_days.append(list_day)
            cur_day += timedelta(days = 1)
        return list_days
        
date_range_2('2018-11-12','2018-10-12')
```

***Задание 4***

В последнем примере поиска по словарю мы использовали 3 столбца. Напишите функцию, которая формирует словарь для поиска по n столбцам.

```python
def dict_to_search(n, path_to_dict):
    stats_dict = {}
    with open(path_to_dict) as f:
        for line in f:
            line = line.strip().split(',')
            if int(n) < int(len(line)):
                stats_dict[tuple(line[:n])] = line[n]
            else:
                return 'Невозможно создать словарь. n - больше длинны исходной строки данных'
    return stats_dict
```

```python
dict_to_search(2,'Python_9_datetime_and_big_files/stats.csv')
```
